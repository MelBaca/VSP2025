```python
import src.DataProcessing.read_data as rd
ruta = '../data/New_tables/LUNES_VIERNES_C4_C6.xlsx'
hoja_c4_c6 = 'C4_C6'
base = rd.tablas_ordinarios_c4_c6_v2(ruta,hoja_c4_c6)

base.head()
```

```python
# Creamos los arcos
import pandas as pd
arcos_costo = []

for i in base.iterrows():
    hora_llegada = i[1]['HORA_LLEGADA']
    a = base[base['HORA_SALIDA']>hora_llegada]
    a = pd.DataFrame(a['HORA_SALIDA'] - hora_llegada).reset_index()
    a['Arco-Costo'] = a.apply(lambda row: (row['index']+1, row['HORA_SALIDA']), axis=1)
    arcos_costo.append(a['Arco-Costo'].tolist())
base['Arcos-Costo']=  pd.Series(arcos_costo)
base['Costo-Inicial'] = 0#base['HORA SALIDA QUITUMBE MINUTOS'] - 300
base

#a[a['HORA SALIDA QUITUMBE MINUTOS']].index

# def creacion_arcos(hora_llegada):
#    quitumbe_l_v['HORA SALIDA QUITUMBE MINUTOS'] <= hora_llegada
    
```

```python
N = base['ID'].unique()
N
```

```python
capacidad = len(base['COD. OP.'].unique())
print(capacidad)
```

```python
costos = {}
for index,row in base.iterrows():
    costos_arcos = {(row['ID'],int(i[0])):i[1] for i in row['Arcos-Costo']}
    costos_arcos[(0,row['ID'])] = row['Costo-Inicial'] 
    costos_arcos[(row['ID'],51)] = 0
    costos.update(costos_arcos)
print(costos)
arcos = costos.keys()
arcos
```

```python
import gurobipy as gp
from gurobipy import GRB

W=1000

m1 = gp.Model('Asignacion-Flota-V1')
# Terminar al alcanzar un Gap del 20%
m1.Params.MIPGap = 0.10
m1.setParam('OutputFlag', 1)
# Enfocar estrategia en mejorar la cota dual
m1.Params.MIPFocus = 3

# Fijar generación muy agresiva de cortes
m1.Params.Cuts = 3

# Terminar luego de 180 segundos
m1.Params.TimeLimit = 1380

x = m1.addVars(arcos,vtype=GRB.BINARY,name='x')
z = m1.addVars(N,vtype=GRB.BINARY,name='z')
# Update model to integrate new variables
m1.setParam('LogFile', 'gurobi.log')

m1.setObjective(gp.quicksum(costos[i,j]*x[i,j] for i,j in arcos)+W*z.sum('*'),GRB.MINIMIZE)

m1.addConstrs((x.sum(n,'*') == 1-z[n] for n in N), name=f"Constraints")
for k in range(1,11):
    m1.addConstrs((x.sum('*',i)-x.sum(i,'*')==0 for i in N ),name=f'conservacion_{k}')

m1.addConstr((x.sum(0,'*') <= capacidad), name=f"Constraints_2")
m1.write('Modelo.lp')

m1.optimize()


```

```python
restricciones = []

if m1.Status == GRB.INFEASIBLE:
    print("Model is infeasible; computing IIS")
    m1.computeIIS()  # Compute the IIS
    m1.write("Modelo.ilp")  # Write the IIS model to a file
    print("The following constraint(s) cannot be satisfied:")
    for c in m1.getConstrs():
        if c.IISConstr:
            print('%s' % c.constrName)
            restricciones.append(c.ConstrName)
restricciones
```
